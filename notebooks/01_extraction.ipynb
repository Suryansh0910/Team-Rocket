```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/ecommerce_customer_data_custom_ratios.csv')
```

```python
print(f"  ROWS    : {df.shape[0]:,}")
print(f"  COLUMNS : {df.shape[1]}")
```

```python
print("\nCOLUMN NAMES & DATA TYPES:")
print(df.dtypes)
```

```python
print("\nFIRST 5 ROWS:")
print(df.head())
```

```python
print("\nMISSING VALUES:")
missing = df.isnull().sum()
missing_pct = (df.isnull().sum() / len(df) * 100).round(2)
print(pd.DataFrame({'Missing Count': missing, 'Missing %': missing_pct}))
```

```python
print(f"\nDUPLICATE ROWS: {df.duplicated().sum()}")
```

```python
print("\nSTATISTICAL SUMMARY:")
print(df.describe())
```

```python
print("\nUNIQUE VALUES PER COLUMN:")
for col in df.columns:
    print(f"  {col:30} → {df[col].nunique():>6} unique")
```

```python
print("\nPRODUCT CATEGORY BREAKDOWN:")
print(df['Product Category'].value_counts())

print("\nPAYMENT METHOD BREAKDOWN:")
print(df['Payment Method'].value_counts())

print("\nGENDER BREAKDOWN:")
print(df['Gender'].value_counts())

print("\nCHURN BREAKDOWN:")
print(df['Churn'].value_counts())
print(f"Churn Rate: {df['Churn'].mean()*100:.2f}%")

print("\nRETURNS BREAKDOWN (excl. nulls):")
print(df['Returns'].value_counts())
```

```python
df['Purchase Date'] = pd.to_datetime(df['Purchase Date'])
print(f"\nDATE RANGE:")
print(f"  Earliest : {df['Purchase Date'].min()}")
print(f"  Latest   : {df['Purchase Date'].max()}")
print(f"  Years    : {sorted(df['Purchase Date'].dt.year.unique())}")
```

```python
print("\nDUPLICATE COLUMN CHECK:")
print(f"'Customer Age' == 'Age': {(df['Customer Age'] == df['Age']).all()}")
print("→ These are identical columns. One will be dropped in cleaning.")
```
